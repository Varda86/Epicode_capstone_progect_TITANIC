# The Titanic: a Data Analysis of Passengers and crew

```python
#useful libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#useful functions

# This function can be used to analyze any DataFrame: it takes each field of the DataFrame
# and calculates their unique values.
# If their number is meaningful it also tells us the frequency of that value in the field.
def find_unique(list_columns, df):
    #These variables can be changed, because they depend on the DataFrame we are analysing
    unique_limit = 50
    frequency = 10
    for col in list_columns:
        unique_val = df[col].nunique()
        if unique_val < unique_limit:
            print(f"\nThe column {col} contains {unique_val} unique values")
            if unique_val < frequency:
                print(f"With this frequency: {df[col].value_counts()}")
                
# Also this function can be used to analyze any DataFrame: it tells us if the fields of the Dataframe
# under investigation contain any null values
def find_null(list_columns, df):
    for col in list_columns:
        if df[col].isna().sum() > 0:
            print(f"The column {col} contains {df[col].isna().sum()} null values")
```

## EXPLORATION OF CORRELATIONS

I want to find a relation between survival rate and other data, or with socio-economic class (represented here by the "class" in wich the passenger traveled) and other data.

### The crew:

```python
Titanic_crew
```

Let's see some quick percentage about gender and survival rate, calculated over the total of Women and Men:

```python
print(f"The total numer of women is {Titanic_crew['gender'].value_counts()['female']}. \n The total numer of men is {Titanic_crew['gender'].value_counts()['male']}")
```

```python
grouped_crew_gender = Titanic_crew.groupby(["gender"])["survived"].value_counts().unstack()
grouped_crew_gender
```

```python
#To see those values in percentages (calculated over the total of females and then of males, because the two numbers are too different):
percentages = grouped_crew_gender.div(grouped_crew_gender.sum(axis=1), axis=0) * 100
percentages
```

As we can see 87% of women could survive, but just 22% of men could.
It's better to quickly show it in a chart, more readable than a table:

```python
percentages.plot(kind="bar", stacked=True, color=["red", "skyblue"])
plt.title("Survival Percentage by Gender")
plt.ylabel("Percentage")
plt.xlabel("Gender")
plt.legend(["Not survived", "Survived"], loc = "upper left")
plt.xticks(rotation=0)

for i in range(len(percentages)):
    for j in range(len(percentages.columns)):
        y_value = percentages.iloc[i, j]
        plt.text(i, y_value / 2 + (percentages.iloc[i, :j].sum() if j > 0 else 0), 
                f'{y_value:.1f}%', 
                ha="center", va="center", color="black")

plt.show()
```

#### Survival Rate

I would like to take a look to some passengers data: I want to find a relation between survival rate and other data, or with socio-economic class (represented here by the "class" in wich the passenger travelled) and other data.

```python
print(f"The total numer of women is {Titanic_passengers['gender'].value_counts()['female']}. \n The total numer of men is {Titanic_passengers['gender'].value_counts()['male']}")
Titanic_passengers.groupby(["gender"])["survived"].value_counts()
```

Clearly there is a correlation between Survival Rate and gender: females had much more possibilities to survive than males.
To better understand it's possible to show the data in a chart:

```python
grouped_psg_gender = Titanic_passengers.groupby(["gender"])["survived"].value_counts().unstack()
percentages = grouped_psg_gender.div(grouped_psg_gender.sum(axis=1), axis=0) * 100

percentages.plot(kind="bar", stacked=True, color=["red", "skyblue"])
plt.title("Survival Percentage by Gender")
plt.ylabel("Percentage")
plt.xlabel("Gender")
plt.legend(["Not survived", "Survived"], loc = "upper left")
plt.xticks(rotation=0)
for i in range(len(percentages)):
    for j in range(len(percentages.columns)):
        y_value = percentages.iloc[i, j]
        plt.text(i, y_value / 2 + (percentages.iloc[i, :j].sum() if j > 0 else 0), 
                f'{y_value:.1f}%', 
                ha="center", va="center", color="black")

plt.show()
```

Now I want to see if there is a connection between class and survival rate:

```python
grouped_psg_class = Titanic_passengers.groupby("class")["survived"].value_counts().unstack()
grouped_psg_class
```

This table shows us how many people survived or not, divided class; data are difficult to read, so like before I take a look at them in percentage (Note that the percentage must be calculated on different quantities, one for each class):

```python
percentages = grouped_psg_class.div(grouped_psg_class.sum(axis=1), axis=0) * 100
percentages
```

```python
percentages.plot(kind="bar", stacked=True, color=["red", "skyblue"])
plt.title("Survival Percentage by Class")
plt.ylabel("Percentage")
plt.xlabel("Class")
plt.legend(["Not survived", "Survived"], loc = "upper left")
plt.xticks(rotation=0)
for i in range(len(percentages)):
    for j in range(len(percentages.columns)):
        y_value = percentages.iloc[i, j]
        plt.text(i, y_value / 2 + (percentages.iloc[i, :j].sum() if j > 0 else 0), 
                f'{y_value:.1f}%', 
                ha="center", va="center", color="black")

plt.show()
```

Looking at the chart we can see that in first class 2/3 of passengers survived, in second class 41% survived, but in third class it was a disaster, about 75% of passengers died.

#### Age of passengers

Now I want to find a relation between "age" field and other data about Passengers.
First I look for a connection between age and gender (notice that I'm using the quantiles, so the number of passengers for each age group is equal):

```python
Titanic_passengers.groupby(pd.cut(Titanic_passengers["age"], [0, 20, 27, 38, 74]))["age"].count()
```

```python
Titanic_passengers.groupby(pd.cut(Titanic_passengers["age"], [0, 20, 27, 38, 74]))["gender"].value_counts().unstack()
```

I dont't see any connection between age and gender, the numbers in the columns are always almost equal, with a lesser presence for women as we know, but nothing changes between the age groups.

Maybe there is a connection between age and Survival pattern:

```python
Titanic_passengers.groupby(pd.cut(Titanic_passengers["age"], [0, 20, 27, 38, 74]))["survived"].value_counts().unstack()
```

There is no connection either between Age and Survived field; the results for each percentile are very similar to each other.

I try with Age and Class:

```python
Titanic_passengers.groupby(pd.cut(Titanic_passengers["age"], [0, 20, 27, 38, 74]))["class"].value_counts().unstack()
```

Finally here we find a connection: while passengers grow older, their number increases in first class and decreases in third class (second class remains almost equal). Well, this is a common pattern that my generation knows very well, in fact we can find this pattern also in our society: older generation is richer than youngers.
Another hypothesis could be that it could have been difficult for older working class people to travel.

It's better to make a chart to display the correlation:

```python
prova = Titanic_passengers.groupby(pd.cut(Titanic_passengers["age"], [0, 20, 27, 38, 74]))["class"].value_counts().unstack()
prova.plot(kind = "line", color=["blue", "green", "red"])
plt.title("Number of Passengers for each Class by Age")
plt.xlabel("Age")
plt.legend(["1st Class", "2nd Class","3rd Class"])
plt.show()
```

N.B. FOR CODERS
In the code below I transform some data, but just because I'm using softwares in italian language, so we use commas for decimals and my Power Query doesn't recognise the dot as decimal separator. If you don't have this kind of problems you don't have to run the code below. I've decided to transform every float in int data, except for the fare, because I don't want to lose the real price of tickets. That's why the CSV file will be separated by a semicolon and not a comma.

```python
Titanic_crew["age"] = Titanic_crew["age"].astype(int)

Titanic_passengers["age"] = Titanic_passengers["age"].astype(int)
Titanic_passengers["ticketno"] = Titanic_passengers["ticketno"].astype(int)
Titanic_passengers["sibsp"] = Titanic_passengers["sibsp"].astype(int)
Titanic_passengers["parch"] = Titanic_passengers["parch"].astype(int)

Titanic_passengers["fare"] = Titanic_passengers["fare"].round(2)
```
